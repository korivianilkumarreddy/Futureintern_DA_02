```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/titanic.csv')
data.head()
```

```python
data = data.drop(columns=['Name','Ticket','Cabin'])
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})

data['Embarked'] = data['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

data.head(5)
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
mode = data.mode().iloc[0]
mode
```

```python
std_dev = data.std()
std_dev
```

```python
median = data.median()
median
```

```python
import matplotlib.pyplot as plt
```

```python
mean_Fare = data['Fare'].mean()
median_Fare = data['Fare'].median()
mode_Fare = data['Age'].mode()[0]


statistics = ['Mean','Median','Mode']
values = [mean_Fare,median_Fare,mode_Fare]


plt.figure(figsize=(8, 5))
plt.bar(statistics, values, color=['blue', 'orange', 'green'])
plt.title('Mean, Median, and Mode of Fare in Titanic Dataset')
plt.ylabel('Fare')
plt.ylim(0, max(values) + 10)
plt.grid(axis='y')


plt.show()
```
